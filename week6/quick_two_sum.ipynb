```python
import bisect
import numpy as np
# load the data
array = np.loadtxt('two_sum.txt')
# sort the array
array.sort()
vis = [0] * 20001

for i, val in enumerate(arr):
    # figure out the position of the smallest value one can have for y for a given x. Use array[i+1:end]
    l = bisect.bisect_left(array, -10000-val, i + 1)
    # figure out the position of the largest value one can have for y for a given x. Use array[i+1:end]
    r = bisect.bisect_right(array, 10000-val, i + 1)
    for j in range(l, r):
        vis[int(val + arr[j] + 10000)] = 1
```
